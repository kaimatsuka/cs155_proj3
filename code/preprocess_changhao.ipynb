# Tokenize

```python
file = open('./data/shakespeare.txt', 'r')
text = file.read()
word_sequence = keras.preprocessing.text.text_to_word_sequence(text, filters='0123456789!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n', lower=True, split=' ')
'''
# treat punctuations (period, comma, colon) as tokens by adding a space in front of them
text = text.replace("."," .")
text = text.replace(","," ,")
text = text.replace(":"," :")

# convert text into word sequence (don't ignore punctuations)
word_sequence = keras.preprocessing.text.text_to_word_sequence(text, filters='0123456789!"#$%&()*+-/;<=>?@[\\]^_`{|}~\t\n', lower=True, split=' ')
'''

Tokenizer = keras.preprocessing.text.Tokenizer(num_words=None, filters='0123456789!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n', lower=True, split=' ', char_level=False, oov_token=None, document_count=0)
Tokenizer.fit_on_texts(word_sequence)
word_counts = Tokenizer.word_counts;
word_index = Tokenizer.word_index
```

# Syllable Dictionary

```python
slb_dict_count = {}
slb_dict_word = {}
with open('data/Syllable_dictionary.txt', 'r') as file:
    for line in file:
        line = line[:-1] # \n
        l = line.split(' ')
        word = l[0]
        slb_dict_word[word] = []
        for i in range(1, len(l)):
            slb_dict_word[word].append(l[i]) #wordsyllable
            if l[i] in slb_dict_count:
                slb_dict_count[l[i]].append(word) #syllablewordgenerate
            else:
                slb_dict_count[l[i]] = [word]
```

```python
print(slb_dict_word)
```

# NLTK Pronouncing Dictionary: pattern of stress

##  Append into Syllable Dictionary: slb_dict

```python
# import nltk
# nltk.download('cmudict')
nltk_dict = nltk.corpus.cmudict.dict()

def slb_count(word):
    count = 0
    slbs = nltk_dict[word][0]
    for i in range(len(slbs)):
        if slbs[i][-1].isdigit():
            count += 1
    return count
```

```python
# test NLTK functionality
print(nltk_dict['abandon'])
print(slb_count('abandon'))
```

```python
for word in word_counts.keys():
    pure_word = word.replace('\'', '') # some words contain \'
    if word not in slb_dict_word.keys() and pure_word in nltk_dict:
        cnt = str(slb_count(pure_word))
        slb_dict_word[word] = cnt
        if cnt not in slb_dict_count.keys():
            slb_dict_count[cnt] = word
        else:
            slb_dict_count[cnt].append(word)
```

# Rhyme Dictionary

```python
# print(slbcnt('adore'))
def rhyme(word):
    temp = nltk_dict[word][0]
    res = ''
    for i in range(len(temp))[::-1]:
        res = temp[i] + res
        if temp[i][-1].isdigit():
            return res
```
